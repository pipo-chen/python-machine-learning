# Matplotlib 库

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## 绘制单条直线

```python
x = np.arange(1,11)
y = 2 * x - 8
plt.plot(x, y, label='test')
plt.legend()
plt.show()
```

## 多条线绘制

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, cosy,label='sin')
plt.plot(x, siny,label='cos')
plt.show()
```

### 1)修改属性 Color

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, cosy, color = "#000000") 
plt.plot(x, siny, color = "c") # r红色 - g 绿色- b 蓝色- y - k 黑色 - w 白色- c青色
plt.show()
```

### 2）修改线条样式

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, cosy, color = "y") 
plt.plot(x, siny, color = "c", linestyle='-')  #-. : -- -
plt.show()
```

### 3) x 轴 y 轴 定义

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, cosy, color = "y")
plt.plot(x, siny, color = "c", linestyle='-')  #-. : -- -
plt.xlabel("x axis")
plt.ylabel('y value')
plt.show()
```

### 4) 标题

```python
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, cosy, color = "y")
plt.plot(x, siny, color = "c", linestyle='-')  #-. : -- -
plt.xlabel("x axis")
plt.ylabel('y value')
plt.title("SIN & COS")
plt.show()
```

## 柱状图

```python
x = [6, 8, 9]
y = [1,2,3] #必须数量一致
plt.bar(x, y, color='r')
plt.show()
```

### 1)绘制多柱状图

```python
x = [6, 8, 9]
y = [1,2,4]
x2 = [9,2,3]
y2 = [4,5,6]

plt.bar(x , y)
plt.bar(x2, y2, color = 'g')

plt.xlabel("price")
plt.ylabel("num")
plt.show() #出现 x y 值一样的两条 -> 覆盖
```

## 散点图

```python
x = np.linspace(0,10,100)
siny = np.sin(x)
cosy = np.cos(x)
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, marker='d')
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha = 0.3) #透明度
plt.show()
```
