## matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100) #起始点 终止点 切成几段
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y) #横坐标对应的值， 纵坐标对应的值 曲线
```

```python
cosy = np.cos(x)
```

```python
cosy
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red")
plt.show() #同时绘制多条曲线 可修改颜色
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.show() #可改变线条样式 -. -- 
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show() #可以设置 x y 轴的范围
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.axis([-1, 11, -2, 2]) #上下左右的空间
plt.show() 
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.xlabel("x axis")
plt.ylabel("x value")
plt.show() #x y轴名称
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color = "red", linestyle = '--', label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("x value")
plt.show() #可以添加线条含义
```

## 散点图 Scatter

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color = "red", linestyle = '--', label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("x value")
plt.legend()
plt.title("Welcome to the ML World!") #可以设置标题
plt.show() 
```

```python
plt.scatter(x, siny)
plt.show()#散点图
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color = "red")
plt.show()#多条
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.3)
plt.show()#可以加透明度
```
