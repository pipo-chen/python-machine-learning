## matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#创建一个 2 * x + 5 的函数
x = np.arange(1,11)
y = 2 * x + 5
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0,10,100) #起始点 终止点 切成几段
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y) #横坐标对应的值， 纵坐标对应的值 曲线
```

```python
cosy = np.cos(x)
```

```python
cosy
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy,color= "red")
plt.show() #同时绘制多条曲线 可修改颜色
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.show() #可改变线条样式 -. -- 
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = ':')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show() #可以设置 x y 轴的范围
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.axis([-1, 11, -2, 2]) #上下左右的空间
plt.show() 
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color = "red", linestyle = '--')
plt.xlabel("x axis")
plt.ylabel("x value")
plt.show() #x y轴名称
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color = "red", linestyle = '--', label="hello")
# plt.xlabel("x axis")
# plt.ylabel("x value")
plt.show() #可以添加线条含义
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_ = 5 * x + 10
y_2 = 2 * x - 8
#建立 subplot 网格，相当于几分之几
plt.subplot(231)

#绘制第一个图像
plt.plot(x, y_sin)
plt.title("Sin")

#绘制第二个图像
plt.subplot(232)
plt.plot(x, y_)
plt.title("FUNC")

#绘制第三个图像
plt.subplot(233)
plt.plot(x, y_2)
plt.title("FUNC_2")

#将第二个 subplot 激活，并绘制第二个图像
plt.subplot(212)
plt.plot(x, y_cos)
plt.title("Cos")

plt.show()
```

## 柱状图

```python
x = np.arange(1,5)
y = np.arange(5,9)

plt.bar(x,y)
plt.show()
```

```python
x = [5,8,10]
y = [12,15,6]
x2 = [5,8,10]
y2 = [6,14,7]

plt.bar(x,y)
plt.ylabel('Y')
plt.xlabel('X')
plt.bar(x2,y2, color='g')

plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color = "red", linestyle = '--', label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("x value")
plt.legend()
plt.title("hello") #可以设置标题
plt.show() 
```

## 散点图 Scatter

```python
x = np.linspace(0,10,100) #起始点 终止点 切成几段
siny = np.sin(x)
cosy = np.cos(x)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y)
plt.show()#散点图
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color = "red")
plt.show()#多条
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.3)
plt.show()#可以加透明度
```
